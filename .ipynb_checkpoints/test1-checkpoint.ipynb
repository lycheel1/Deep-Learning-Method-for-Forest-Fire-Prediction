```python
df1.loc[:, ["SDATE", "EDATE"]].max(axis=1)
from datetime import date
zero = date(1900,1,1)
df1.loc[:,["SDATE", "EDATE"]].fillna(zero)
```

```python
df1.loc[:, "sDate"] = df1.loc[:, ["SDATE","AFSDATE", "EDATE", "AFEDATE"]].min(axis=1)
df1.loc[:, "eDate"] = df1.loc[:, ["SDATE","AFSDATE", "EDATE", "AFEDATE"]].max(axis=1)
```

```python
df1.loc[:,["SDATE","EDATE","AFSDATE","AFEDATE","sDate","eDate"]]
```

```python
df1.loc[:,"days"] = df1.loc[:,"eDate"] - df1.loc[:, "sDate"]
```

```python
import numpy as np
np.sum(df1.loc[:, "days"].isna()) 
print(df1.loc[df1.loc[:,"days"].isna()].index.tolist())
df1 = df1.drop(columns=["SDATE", "EDATE", "AFSDATE", "AFEDATE"])
df1 = df1.dropna(axis=0, how='any')
```
